```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
# scatter plot with a linear fit on top of that
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,
           hue='sex',markers=['o','v'])
# marker styles come from Matplotlib
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,
           hue='sex',markers=['o','v'],
          scatter_kws={'s':100})
# the dictionary scatter_kws={'s':100} affects marker sizes
# need to refer documentation for any needed specific case
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,
           col='day',row='time',hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,
           col='day',hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,
           col='day',hue='sex',
          aspect=0.6,size=8)
```
