```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# this aloows us to see the plots we create 
# right inside the jupyter notebook
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
# 0 to 5, 11 points
y = x**2 # x^2
```

```python
x
```

```python
y
```

```python
# Functional method
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.show()
# have to type plt.show() if not using
# jupyter notebook
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# Multiplots
```

```python
plt.subplot(1,2,1)
# number of rows, number of columns
# and a number of plot we are refering to
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')

# Have two subplots
```

```python
# Now object oriented method (OO)
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)

axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Set Title')
```

```python
# create empty canvas
# then add set of axes

fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
# [left, bottom, width, height] 

axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('LARGER PLOT')
axes2.set_title('SMALLER PLOT')
```

```python
# Create subplot with OO
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#axes.plot(x,y)
```

```python
axes
# axes is array of axes
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#axes.plot(x,y)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#
# comment on this "comma" syntax:
#  we're using tuple unpacking here 
# to grab the entire figure object and then an axes object, 
# which is a list of all the axes available to us
#
axes[0].plot(x,y)
axes[0].set_title('First plot')
axes[1].plot(y,x)
axes[1].set_title('Second plot')
```

```python
# Figure size and DPI (dots per inch)
```

```python
# fig = plt.figure(figsize=(3,2),dpi=100)

# Let's use default DPI:

fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(figsize=(8,2))

axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
# Add legend

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')

ax.legend(loc=0)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')

ax.legend(loc=10)
# loc = 10
# location code corresponding to 'center'
# list of location code available at matplotlib documantation
# https://matplotlib.org/api/legend_api.html
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')

ax.legend(loc=(0.1,0.1))
```
